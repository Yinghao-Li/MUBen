```python
import lmdb
import numpy as np
import os
import pickle
import torch
import numpy as np
```

```python
lmdb_path = "../data/UniMol/tox21/test.lmdb"

env = lmdb.open(
    lmdb_path,
    subdir=False,
    readonly=True,
    lock=False,
    readahead=False,
    meminit=False,
    max_readers=256,
)
txn = env.begin()
keys = list(txn.cursor().iternext(values=False))
ori_ids = list()
lbs = list()
for idx in keys:
    datapoint_pickled = txn.get(idx)
    data = pickle.loads(datapoint_pickled)
    ori_ids.append(data['ori_index'])
    lbs.append(data['target'])
```
